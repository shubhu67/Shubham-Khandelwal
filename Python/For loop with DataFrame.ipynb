```python
# importing pandas
import pandas as pd
```

```python
# Creating new dataframe
initial_data = {'First_name': ['Ram', 'Mohan', 'Tina', 'Jeetu', 'Meera'],
                'Last_name': ['Kumar', 'Sharma', 'Ali', 'Gandhi', 'Kumari'],
                'Marks': [12, 52, 36, 85, 23] }
 
df = pd.DataFrame(initial_data, columns = ['First_name', 'Last_name', 'Marks'])
```

```python
# Generate result using pandas
result = []
for value in df['Marks']:
    if value >= 33:
        result.append("Pass")
    elif value < 0 and value > 100:
        result.append("Invalid")
    else:
        result.append("Fail")
      
df["Result"] = result  
df
```

```python
for i in df['Marks']:
    print(i)
```

###### OR

```python
df['Results'] = ['Pass' if m>=33 else 'Fail' for m in df['Marks']]
df
```

```python
df
```

```python
df['rank_marks']=df['Marks'].rank(method='dense', ascending=False)
df
```

```python
df_main=df[df['rank_marks']<3]
```

```python
df_main.sort_values(by='rank_marks', inplace=True, ignore_index=True)
```
