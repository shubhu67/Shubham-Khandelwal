##### Relational & Categorical Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales=pd.read_excel('sales.xlsx')
sales.sample(5)
```

```python
sales.columns
```

```python
sns.relplot(x='Sales', y='Profit', data=sales)
plt.title('Profit vs Sales')
```

```python
# hue parameter adds additional parameter to our visual 
sns.relplot(x='Sales', y='Profit', data=sales, hue='Order Priority')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, hue='Order Priority', style='Ship Mode')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, size='Discount')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, size='Discount', sizes=(25, 100))
```

```python
sns.relplot(x='Order Date', y='Sales', data=sales, kind='line')
```

```python
sales['year']=sales['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', ci=None)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', ci='sd')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', estimator=None)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', estimator='sum')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', style='Product Category')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', style='Product Category', ci=None)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', col='Product Category', ci=None)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', col='Region', ci=None)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', col='Region',col_wrap=4, ci=None)
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment', col_wrap=2)
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment', col_wrap=2, hue='Order Priority')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment', col_wrap=2, hue='Ship Mode')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment', row='Product Category', hue='Ship Mode')
```

* Strip & Swarm plots represent an implementation of a scatterplot for a categorical variable

```python
sns.catplot('Ship Mode', 'Sales', data=sales, kind='strip')
```

```python
sns.catplot('Ship Mode', 'Sales', data=sales, jitter = False)
```

```python
sns.catplot('Ship Mode', 'Sales', data=sales.query('Profit>1000'), jitter = False)
```

```python
#  To avoid overlapping of data points, we use swarm plot
sns.catplot('Ship Mode', 'Sales', data=sales.query('Profit>1000'), kind='swarm')
```

```python
#  The median shipping cost for furniture is very high
sns.catplot(x='Product Category', y='Shipping Cost', data=sales, kind='box')
```

```python
# The shipping cost in case of deliveries by truck is always high
sns.catplot(x='Product Category', y='Shipping Cost',hue='Ship Mode', data=sales, kind='box')
```

```python
# The shipping cost in case of deliveries by truck is always high
sns.catplot(x='Product Category', y='Shipping Cost',hue='Ship Mode',hue_order=['Delivery Truck', 'Regular Air', 'Express Air'], data=sales, kind='box')
```

```python
# The shipping cost in case of deliveries by truck is always high
sns.catplot(x='Product Category', y='Shipping Cost',row='Ship Mode', data=sales, kind='box')
```

```python
sns.catplot(x='Product Category', y='Shipping Cost',col='Ship Mode', data=sales, kind='box')
```

```python
sns.catplot(x='Product Category', y='Shipping Cost',col='Ship Mode',row='Customer Segment', data=sales, kind='box')
```

* Violin plot is like a mixture of both the Box plot and the density plot. It illustrates the shape of the kernel density estimate to capture the spread of the data, along with the important visual description from the box plot.



```python
sns.catplot(x='Product Category', y='Shipping Cost', data=sales, kind='violin')
```

```python
sns.catplot(x='Product Category', y='Shipping Cost', hue='Ship Mode',data=sales, kind='violin')
```

```python
sns.catplot(x='Product Category', y='Sales',col='Ship Mode', data=sales, kind='bar')
```

```python
sns.catplot(x='Product Category', y='Sales',hue='Ship Mode', data=sales, kind='bar')
```

```python
sns.catplot(x='Product Category', y='Sales',hue='Ship Mode',row='Customer Segment' ,data=sales, kind='bar')
```
