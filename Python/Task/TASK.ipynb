#### OR

```python
task1['flag']=task1['body'].apply(lambda x:1 if 'solar' in x.lower().split() else 0)
```

```python
task1['flag'].value_counts().loc[1]
```

### OR

```python
def solar(x):
    flag=0
    for i in x.lower().split():
        if i=='solar':
            flag=1
            break
    return(flag)
```

```python
solar('solar panel')
```

```python
solar('solr panel')
```

```python
task1['flag']=task1['body'].apply(solar)
task1
```

```python
task1['flag'].value_counts(ascending=True).iloc[1]
```

### 5.Add 3 columns and extract the quarter, month and year from date.

```python
task1['Quarter']=task1['date'].dt.quarter
task1['Month']=task1['date'].dt.month
task1['Year']=task1['date'].dt.year
task1.head()
```
