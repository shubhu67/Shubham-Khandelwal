# If-then/If-then-else

```python
# 1) Assignment to one or more columns using If-then/If-then-else on some specific column
# 2) Suppose we have three columns A, B and C, assume depends on some value in A column, wanted to change the value in B & C
# 3) With the help of loc function, we will put the condition on column A, and assign the value in B and C
# 4) Code snippet
# Creating the dataframe
df = pd.DataFrame({'A': [4, 5, 6, 7],
                   'B': [10, 20, 30, 40],
                    'C': [100, 50, -30, -50]})
# Using loc funtion we are putting the condition If-then on the column A, and Assigning new value in Column B,
# when Column A is having Greater than 5
df.loc[df.A >= 5, 'B'] = 'New Value'
df
```

```python
# In the case of If-then-else
df = pd.DataFrame({'A': [4, 5, 6, 7],
                   'B': [10, 20, 30, 40],
                    'C': [100, 50, -30, -50]})
# Using loc funtion we are putting the condition If-then-else on the column A, and Assigning new value in Column B and C
# when Column A is having Greater than 5 else Lesser than that will remain the another value in Column B and C 
df.loc[df.A >= 5, ['B','C']] = 'Greater than 5'
df
```

```python
df.loc[df.A < 5, ['B','C']] = 'Lesser than 5'
df
```

```python
# The another way we could have encounter whether coulmn A is having greater than 5 or not
df = pd.DataFrame({'A': [4, 5, 6, 7],
                   'B': [10, 20, 30, 40],
                   'C': [100, 50, -30, -50]})
df['logic'] = np.where(df['A'] > 5, 'high', 'low')
df
```

```python
# 1) Grouping of the data using sub data dataframe
# 2) Grouping required for segregating, but we might need to group on another sub dataframe as well 
# 3) This can be achieve by utilizing the lambda function and finding the maximum values, under which we can use sub data frame
# 4) Code snippet
```

```python
 df = pd.DataFrame({'animal': 'cat dog cat fish dog cat cat'.split(),
                    'size': list('SSMMMLL'),
                    'weight': [8, 10, 11, 1, 20, 12, 12],
                    'adult': [False] * 5 + [True] * 2})
df
```

```python
# List the size of the animals with the highest weight.
df.groupby('animal').apply(lambda subf: subf['size'][subf['weight'].idxmax()])
```

```python
# the same can be used by get_group also
gb = df.groupby(['animal'])
gb.get_group('cat')
```

```python
# In this function we are trying to calculate the average weight by the following and the using apply method calling the same
# Finding the weight whose size is S and multiplying with 1.5, same mechanism for all remaing element and storing is same variabl
# and then dividing by the len of the elements
def GrowUp(x):
        avg_weight = sum(x[x['size'] == 'S'].weight * 1.5)
        avg_weight += sum(x[x['size'] == 'M'].weight * 1.25)
        avg_weight += sum(x[x['size'] == 'L'].weight)
        avg_weight /= len(x)
        return pd.Series(['L', avg_weight, True],
                         index=['size', 'weight', 'adult'])
```

```python
expected_df = gb.apply(GrowUp)
expected_df
```

# Pandas Shift Function

The function can take up to 4 parameters.
1) Periods by default is equal to 1 indicates the number of periods to shift. It can be positive or negative. A negative number will shift the values up.
2) freq is to preserve the original data; that means that the index is not shifting only the data.
3) There is an axis parameter to tell the direction of the shift. The possible values are 0 or index, 1 or columns.
4) fill_value can be used for the missing values introduced as a result of the shifting. By moving the data, we will miss some values for either the first index or the last  depends on the period value. 
We can decide to fill those NA by a value of our choice.

```python
df=pd.DataFrame({'A': [10,20,30,40,50],
                 'B': [11,21,31,41,51],
                 'C': [45,90,135,180,225]})
df.head()
```

What is the outcome of shifting by 1 period?

```python
df.shift(periods=1)
```

```python
# Shifting value for a series & saving to another column
```

```python
df['A_shift']=df['A'].shift(1)
df
```

The value of the period indicates how to move the data. In the first case, the output resulted in all the columns shifted. The value initially in index 0 shifted to index 1 and so on. Also, we can shift for only one column. The result can be saved in a new column for future use. Note that we lose the last value and the first value is NaN.

```python
# Output of  a negative  period value
df.shift(-1)
```

How does the result change by applying the axis?

```python
df.shift(periods=1,axis='columns')
```

```python
df.shift(periods=1,axis='index')
```

By default the axis is set to the index value. The second line of code returned the same output that we had earlier. However, if we want to shift the value of the dataframe to the next columns, we have to pass columns as value for axis.

How to remove the NaN?

```python
df=pd.DataFrame({'A': [10,20,30,40,50],
                 'B': [11,21,31,41,51],
                 'C': [45,90,135,180,225]})
df.head()
# use of fillna
df.shift(periods=2,fill_value=0)
```

```python
df['C_shift']=df['C'].shift(periods=2,fill_value=10)
df
```

The fill_value parameter is important as it helps us remove the NAs and based on the project, we will replace all the NaN by the best value. Some models dont take null values and in such cases we will have to either drop the NAs or replace them by the appropriate value. Note that the value given will replace all the NAN that we should have otherwise.

# Topic 15 : Calculating common elements from Dataframe

```python
# Calculate common elements in a df depending on another column ?

# wanted to calculate the number of sports shared by two countries. For a example
# Football and Volleyball is common to brazil and Uk. So the number of common sports played by brazil and Uk is 2.


from itertools import combinations, product

df = pd.DataFrame({'sport': ['Foot_ball','Foot_ball','Volleyball','Volleyball','Volleyball','Rugger','Rugger','Rugger','Carrom','Carrom','Chess','Chess'],
                   'country': ['brazil','UK','UK','South_Africa','brazil','UK','South_africa','Australia','UK','Australia','UK','Australia']})
df

# Fix Capitalization
df['country'] = ['_'.join(map(str.title, x.split('_'))) for x in df['country']]

c0, c1 = zip(*[(a, b)
               for s, c in df.groupby('sport')['country']
               for a, b in combinations(c, 2)])

i, r = pd.factorize(c0)
j, c = pd.factorize(c1)
n, m = len(r), len(c)
o = np.zeros((n, m), np.int64)
np.add.at(o, (i, j), 1)

result = pd.DataFrame(o, r, c)
result

#               Australia  Uk  South_Africa  Brazil
# Uk                    3   0             2       1
# Brazil                0   1             0       0
# South_Africa          1   0             0       1
```

```python
c0, c1 = map(pd.Series, zip(*[(a, b)
   for s, c in df.groupby('sport')['country']
   for a, b in combinations(c, 2)]))

pd.crosstab(c0, c1).rename_axis(None).rename_axis(None, axis=1).pipe(
    lambda d: d.align(d.T, fill_value=0)[0]
)
```

# Topic 19 : Update values all over row diagonally

```python
def fill_wrapped_diag(a, fill_val):
#     this will get the shape of dataframe i.e. 7 and 3
    r,c = a.shape
#     Logical part to identify for shifting on the left side such 3-7%3 output will be 2
    r_left = c-r%c    
    a_ext = np.pad(a, ((0,r_left),(0,0)))
#     Redesigning the shape again by identifying how many palce we need to shift based on r, r_left and c
    a_r = a_ext.reshape((r+r_left)//c, -1)
#     From the position + 1 we will substitute the fill_val parameter
    a_r[:,::c+1] = fill_val
    return a_r.reshape(a_ext.shape)[:-r_left]


# Creating a dataframe with 7*3 matrix by inserting random value between 0 to 10
a = np.random.randint(0,10,(7,3))
# Calling the function, and passing the dataframe, and the updated values
fill_wrapped_diag(a, fill_val=75)
```

# Topic 39 : Fill missing data in specified range

```python
# Creating the dataframe by keeping the time and date values in it
df = pd.DataFrame({'day': ['2012-10-12', '2012-10-12', '2012-10-12'],
                   'min_from': [0,5,6],
                  'volume':[0,0,0]})
# As Labelled Index
# df = df.set_index('day')
df
```

```python
# In case day is not datetime yet:
df['Date'] = pd.to_datetime(df['day']) + pd.to_timedelta(df['min_from'],unit='m')
df = df.set_index('Date')
df
```

```python
# df['Date'] = df.index.date
sample_days = df['Date'].unique()

# Resample to 1 minute and keep only the original dates
df = df.resample('1t').last()
df = df.loc[df['Date'].isin(sample_days)]

# Remove non open hours
df = df.between_time('00:00', '15:30')

# Forward fill the remaining columns (notice, as NAs in volume are removed, it does effect this column)
df = df.fillna(method='ffill')
df
```

# Topic 40 : How to use apply function on existing columns with global variables as additional arguments?

In df, use apply method to replace the missing values in Min.Price with the columns mean and those in Max.Price with the columns median.
The args argument given to apply function is passed func argument (lambda function given).

```python
# Creating the dataframe with two columns
import pandas as pd
import numpy as np
df = pd.DataFrame({'Manufacturer': ['Audi','BMW','Chevrolet'],
                   'Model': ['90','535i','Cavalier'],
                   'Min.Price': [50,60,70],
                   'Max.Price': [80,100,120]})
df
```

```python
d = {'Min.Price': np.nanmean, 'Max.Price': np.nanmedian}
df[['Min.Price', 'Max.Price']] = df[['Min.Price', 'Max.Price']].apply(lambda x, d: x.fillna(d[x.name](x)), args=(d, ))
```

# Topic 41 : How to change the order of columns of a dataframe?

Answering 3 questions here :

a) In df, interchange columns 'a' and 'c'.

b) Create a generic function to interchange two columns, without hardcoding column names.

c) Sort the columns in reverse alphabetical order, that is colume 'e' first through column 'a' last.

```python
# Input : 
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
df
```

```python
#Solution
# Solution Q a)
df[list('cbade')]
```

```python
# Solution Q b)
def switch_columns(df, col1=None, col2=None):
    colnames = df.columns.tolist()
    i1, i2 = colnames.index(col1), colnames.index(col2)
    colnames[i2], colnames[i1] = colnames[i1], colnames[i2]
    return df[colnames]

df1 = switch_columns(df, 'a', 'c')

df1
```

```python
# Solution Q c)
df[sorted(df.columns)]
```

# Topic 42 : How to reshape a dataframe to the largest possible square after removing the negative values?

Reshape df to the largest possible square with negative values removed. Drop the smallest values if need be. The order of the positive numbers in the result should remain the same as the original.

```python
# Input
df = pd.DataFrame(np.random.randint(-20, 50, 100).reshape(10,-1))
df
```

```python
# Step 1: remove negative values from arr
arr = df[df > 0].values.flatten()
arr_qualified = arr[~np.isnan(arr)]

# Step 2: find side-length of largest possible square
n = int(np.floor(arr_qualified.shape[0]**.5))

# Step 3: Take top n^2 items without changing positions
top_indexes = np.argsort(arr_qualified)[::-1]
output = np.take(arr_qualified, sorted(top_indexes[:n**2])).reshape(n, -1)
print(output)
```

# Topic 43: How to create a new column that contains the row number of nearest column by euclidean distance?

Create a new column such that, each row contains the row number of nearest row-record by euclidean distance.

```python
# Input : 
df = pd.DataFrame(np.random.randint(1,100, 40).reshape(10, -1), columns=list('pqrs'), index=list('abcdefghij'))
df
```

```python
#-------------------------------------
# Desired output
#------------------------------------

#    p   q   r   s nearest_row   dist
# a  57  77  13  62           i  116.0
# b  68   5  92  24           a  114.0
# c  74  40  18  37           i   91.0
# d  80  17  39  60           i   89.0
# e  93  48  85  33           i   92.0
# f  69  55   8  11           g  100.0
# g  39  23  88  53           f  100.0
# h  63  28  25  61           i   88.0
# i  18   4  73   7           a  116.0
# j  79  12  45  34           a   81.0
```

```python
# Solution
import numpy as np

# init outputs
nearest_rows = []
nearest_distance = []

# iterate rows.
for i, row in df.iterrows():
    curr = row
    rest = df.drop(i)
    e_dists = {}  # init dict to store euclidean dists for current row.
    # iterate rest of rows for current row
    for j, contestant in rest.iterrows():
        # compute euclidean dist and update e_dists
        e_dists.update({j: round(np.linalg.norm(curr.values - contestant.values))})
    # update nearest row to current row and the distance value
    nearest_rows.append(max(e_dists, key=e_dists.get))
    nearest_distance.append(max(e_dists.values()))

df['nearest_row'] = nearest_rows
df['dist'] = nearest_distance
```

# Topic 46: How to get the items not common to both series A and series B?

Get all items of ser1 and ser2 not common to both.

```python
# Input
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

```python
# Solution
ser_u = pd.Series(np.union1d(ser1, ser2))  # union
ser_i = pd.Series(np.intersect1d(ser1, ser2))  # intersect
ser_u[~ser_u.isin(ser_i)]
```

# Topic 47: How to find and cap outliers from a series or dataframe column?

Replace all values of ser in the lower 5%ile and greater than 95%ile with respective 5th and 95th %ile value.

```python
# Input
ser = pd.Series(np.logspace(-2, 2, 30))
```

```python
# Solution
def cap_outliers(ser, low_perc, high_perc):
    low, high = ser.quantile([low_perc, high_perc])
    print(low_perc, '%ile: ', low, '|', high_perc, '%ile: ', high)
    ser[ser < low] = low
    ser[ser > high] = high
    return(ser)

capped_ser = cap_outliers(ser, .05, .95)
```
