```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
  'restaurant_id': [101,102,103,104,105,106,107],
  'address': ['A','B','C','D', 'E', 'F', 'G'],
  'city': ['London','London','London','Oxford','Oxford', 'Durham', 'Durham'],
  'sales': [10,500,48,12,21,22,14]
})
```

```python
df
```

```python
grouped=df.groupby('city')[['sales']].agg({'sum'})#.rename(columns={'sales':'city_total_sales'}).reset_index()
grouped
```

```python
merged=pd.merge(df, grouped, how='left', left_on='city', right_on='city')
merged
```

```python
merged['pct']=(merged['sales']/merged['city_total_sales']*100).round(2)
merged
```
